### Data preparation

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
number_cat = 10
```

```python
y_train
```

```python
# convert y_train to binary values
import keras
y_train = keras.utils.to_categorical(y_train, number_cat)
```

```python
y_train
```

```python
y_test = keras.utils.to_categorical(y_test, number_cat)
y_test
```

```python
X_train
```

```python
#normalise to have it between 0-1
# more resolution within our data points
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
X_train.shape
```

```python
Input_shape = X_train.shape[1:]
```

```python
Input_shape
```

### Train model

```python
#We are going to build our model in sequential manner
from keras.models import Sequential

#Conv2D- For Convolution Process
#MaxPooling2D - For performing max pooling that reduces the dimensionality of the image and kind of compressing actual features
#AverageCooling2D - For performing average function while pooling(like how max function takes maximum value, avg will take avg value)
#Dense - To create fully connected ANN
#Flatten - To flatten our array from 2d to 1d
#Dropout - Perform regularisation which is dropping of the neurons along with associated weights to improve the generalisation capability of the network
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout

# To obtain weights
from keras.optimizers import Adam

#For callback to keras API
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
#32 filters, 3x3 kernel matrix
cnn_model.add(Conv2D(filters = 128, kernel_size = (3, 3), activation = 'relu', input_shape = Input_shape))
cnn_model.add(Conv2D(filters = 128, kernel_size = (3, 3), activation = 'relu'))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Dropout(0.3))

cnn_model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu'))
cnn_model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Dropout(0.2))

cnn_model.add(Flatten())

#units is how many neurons i need
cnn_model.add(Dense(units = 512, activation = 'relu'))
cnn_model.add(Dense(units = 512, activation = 'relu'))

#output layer
#softmax is a simple function used for classification purposes
cnn_model.add(Dense(units = 10, activation = 'softmax'))
```

```python
#keras.optimizers.RMSprop(lr = 0.001)
cnn_model.compile(loss = 'categorical_crossentropy', optimizer = keras.optimizers.RMSprop(lr = 0.001), metrics = ['accuracy'])
```

```python
#History is an object or a variable they're going to be capturing all the progression of the network
#as it is being trained
#shuffle so that input is random
history = cnn_model.fit(X_train, y_train, batch_size = 32, epochs = 1, shuffle = True)
```

### Evaluate the model

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy: {}'.format(evaluation[1]))
```
